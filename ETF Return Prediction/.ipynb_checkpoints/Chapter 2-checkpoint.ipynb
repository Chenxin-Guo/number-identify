### 2.4

Recall that $N(\beta, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x - \beta)^2}{2\sigma^2}}$, $f(y, \theta_i) = (2\pi\sigma_i^2)^{-\frac n2}e^{-\frac{\Sigma_j(y_j - \beta_i)^2}{2\sigma_i^2}}$

$I(\theta_1;\theta_2) = \frac1n \mathbb{E}_1log[((\frac{2\pi\sigma_1^2}{2\pi\sigma_2^2})^{-\frac n2}\frac{-\frac{\Sigma_j(y_j - \beta_1)^2}{2\sigma_1^2}}{-\frac{\Sigma_j(y_j - \beta_2)^2}{2\sigma_2^2}}]\\
=\frac1n( \mathbb{E}_1 log(\frac{\sigma_1^2}{\sigma_2^2})^{-\frac n2} + \mathbb{E}_1(\frac{\Sigma_j(y_j - \beta_2)^2}{2\sigma_2^2} - \frac{\Sigma_j(y_j - \beta_1)^2}{2\sigma_1^2}))\\
=-\frac12log(\frac{\sigma_1^2}{\sigma_2^2}) + \frac1n\frac{\Sigma_j\mathbb{E}_1(y_j^2 - 2y_j\beta_2 + \beta_2^2)}{2\sigma_2^2}- \frac1n\frac{\Sigma_j\mathbb{E}_1(y_j^2 - 2y_j\beta_1 + \beta_1^2)}{2\sigma_1^2}$

Since $Var(x) = \mathbb{E}(x^2) - (\mathbb{E}x)^2$, for any $j$, $\mathbb{E}_1(y_j) = \beta_1$, $\mathbb{E}_1(y_j^2) = \sigma_1^2 + \beta_1^2$

$I(\theta_1;\theta_2) = -\frac12log(\frac{\sigma_1^2}{\sigma_2^2}) + \frac1n\frac{\Sigma_j(\sigma_1^2 + \beta_1^2 - 2\beta_1\beta_2 + \beta_2^2)}{2\sigma_2^2} - \frac1n\frac{\Sigma_j(\sigma_1^2 + \beta_1^2 - 2\beta_1^2 + \beta_1^2)}{2\sigma_1^2}\\
=-\frac12log(\frac{\sigma_1^2}{\sigma_2^2}) + \frac1n\frac{\Sigma_j \sigma_1^2 + \Sigma_j(\beta_1 - \beta_2)^2}{2\sigma_2^2} - \frac12\\
=-\frac12log(\frac{\sigma_1^2}{\sigma_2^2}) + \frac{\sigma_1^2}{2\sigma_2^2} + \frac1n\frac{\Sigma_j(\beta_1 - \beta_2)^2}{2\sigma_2^2} - \frac12\\
=\frac12(\frac{\sigma_1^2}{\sigma_2^2} - log(\frac{\sigma_1^2}{\sigma_2^2}) - 1) + \frac{(\beta_1 - \beta_2)^TZ^TZ(\beta_1 - \beta_2)}{2n\sigma_2^2}$

### 2.6

#### (a)

$\mathbb{E}(x_t) = \beta_0 + \beta_1 t$, which depends on $t$.

#### (b)

$\nabla x_t = x_t - x_{t-1} = \beta_0 - \beta_0 + \beta_1t - \beta_1(t - 1) + w_t - w_{t-1} = \beta_1 + w_t - w_{t-1}$

$\mathbb{E}(x_t - x_{t-1}) = \beta_1$, which does not depends on $t$.

$\gamma_{\nabla x_t}(s, t) = \mathbb{E}[(\beta_1 + w_t - w_{t-1} - \beta_1)(\beta_1 + w_s - w_{s-1} - \beta_1)] = \mathbb{E}(w_tw_s - w_tw_{s-1} - w_{t-1}w_s + w_{t-1}w_{s-1})\\
=\begin{cases} 
      2 & |s - t| = 0\\
      -1 & |s - t| = 1\\
      0 & |s - t| \geq 2\\
\end{cases}$

, which only depends on $h = |s-t|$. Therefore $\nabla x_t$ is stationary.

#### (c)

Assume the mean of $y_t$ is $\mu_y$ and the ACF is $\gamma_y(h)$.

$\nabla x_t = x_t - x_{t-1} = \beta_0 - \beta_0 + \beta_1t - \beta_1(t - 1) + y_t - y_{t-1} = \beta_1 + y_t - y_{t-1}$

$\mathbb{E}(x_t - x_{t-1}) = \beta_1 + \mu_y - \mu_y = \beta_1$, which does not depends on $t$.

$\gamma_{\nabla x_t}(s, t) = \mathbb{E}[(\beta_1 + y_t - y_{t-1} - \beta_1)(\beta_1 + y_s - y_{s-1} - \beta_1)] = \mathbb{E}(y_ty_s - y_ty_{s-1} - y_{t-1}y_s + y_{t-1}y_{s-1})$.

Recall that $\mathbb{E}(y_ty_s) = \gamma_y(|s-t|) - \mu_y^2$.

$\gamma_{\nabla x_t}(s, t) = 2\gamma_y(h) - \gamma_y(h+1) - \gamma_y(h-1) + 3\mu_y^2$, which only depends on $h = |s-t|$. Therefore $\nabla x_t$ is stationary.



### 2.7

Assume the mean of $y_t$ is $\mu_y$ and the ACF is $\gamma_y(h)$.

$\mathbb{E}(x_t - x_{t-1}) = \delta + \mathbb{E}(y_t) - \mathbb{E}(y_{t-1}) = \delta$

$\gamma(s, t) = \mathbb{E}[(\delta + w_t + y_t - y_{t-1} - \delta)(\delta + w_2 + y_2 - y_{2-1} - \delta)]\\
=\mathbb{E}[(w_t + y_t - y_{t-1})(w_s + y_s - y_{s-1})]\\
=\mathbb{E}(w_tw_s + y_tw_s -y_{t-1}w_s + w_ty_s + y_ty_s - y_{t-1}y_s - w_ty_{s-1} - y_ty_{s-1} + y_{t-1}y_{s-1})$

Since $w_t$ is independent to $y_t$.

$\gamma(s, t)=\mathbb{E}(w_tw_s + y_ty_s - y_{t-1}y_s - y_ty_{s-1} + y_{t-1}y_{s-1})\\
=\begin{cases} 
      2\gamma_y(h) - \gamma_y(h+1) - \gamma_y(h-1) + 3\mu_y^2 + 1 & |s - t| = 0\\
      2\gamma_y(h) - \gamma_y(h+1) - \gamma_y(h-1) + 3\mu_y^2 & |s - t| \neq 0\\
\end{cases}$

, which only depends on $h = |s-t|$. Therefore (2.27) is stationary.

#### (a)

```python
data = pd.read_csv('./data/oil.csv').merge(pd.read_csv('./data/gas.csv'))
plot = sns.lineplot(x='date', y='value', hue='variable', data=data.melt(id_vars='date', value_vars=['oil', 'gas']))
```

#### (b)

$y_t$ can be interpreted as the change in maginitude of price.

#### (c)

```python
data = pd.read_csv('./data/oil.csv').merge(pd.read_csv('./data/gas.csv'))
data['oil'], data['gas'] = np.log(data['oil']).diff(1), np.log(data['gas']).diff(1)
data = data.drop(0).reset_index(drop=True)
plot = sns.lineplot(x='date', y='value', hue='variable', data=data.melt(id_vars='date', value_vars=['oil', 'gas']))
```

```python
mu = np.mean(data['oil'])
gamma_oil = [sum([(data['oil'][i] - mu)*(data['oil'][i+lag]-mu) for i in range(len(data['oil']) - lag)]) / len(data['oil']) for lag in range(100)]
rho_oil = [item / gamma_oil[0] for item in gamma_oil]
mu = np.mean(data['gas'])
gamma_gas = [sum([(data['gas'][i] - mu)*(data['gas'][i+lag]-mu) for i in range(len(data['gas']) - lag)]) / len(data['gas']) for lag in range(100)]
rho_gas = [item / gamma_gas[0] for item in gamma_gas]
plot = sns.lineplot(x=list(range(100)) * 2, y=rho_oil + rho_gas, hue=['oil'] * 100 + ['gas'] * 100)
```

After the transformation, two time series behave like stationary time series.

#### (d)

```python
mu_oil = np.mean(data['oil'])
mu_gas = np.mean(data['gas'])
gamma = [sum([(data['oil'][i] - mu_oil)*(data['gas'][i+lag]-mu_gas) for i in range(len(data['date']) - lag)]) / len(data['date']) for lag in range(100)]
rho = [item / math.sqrt(gamma_oil[0] * gamma_gas[0]) for item in gamma]
plot = sns.lineplot(x=range(100), y=rho)
```

When $4 > h > 0$, $\rho(h)$ is positive, which means gas is leads oil a few weeks.

#### (e)

```python
for i in range(1, 4): data['oil_' + str(i)] = data['oil'].shift(i)
plot = sns.pairplot(data[['gas', 'oil_1', 'oil_2', 'oil_3']][3:], kind="reg")
```

We can see some outliers as the edge of each plot. And when the time shift is 1 week, the linear relationship is stronger.

#### (f)

```python
data['indicator'], data['intercept'] = (data['oil'] >= 0).astype('int'), 1
data = data[1:][['gas', 'intercept', 'indicator', 'oil', 'oil_1']]
model = sm.OLS(data['gas'], data.drop(['gas'], axis=1)).fit()
model.summary()
```

When there is a negative growth, the model is $G_t = -0.0065 + 0.6809O_t + 0.1115O_{t-1} + w_t$.

When there is no or positive growth, the model is $G_t = 0.0062 + 0.6809O_t + 0.1115O_{t-1} + w_t$.

The result shows that the intercept is very close in the two case, thus does not support the asymmetry hypothesis.

```python
plot = sns.scatterplot(x=data['gas'], y=model.predict(data.drop(['gas'], axis=1)))
```

we can observe a high-leverage point and the residuals plot indicates a linear relationship.

### 2.11

```python
data = pd.read_csv('./data/globtemp.csv')
data['moving_average'] = np.mean(np.concatenate([[data['globtemp'].shift(i) for i in range(-5, 6)]]), axis=0)
data['kernel_smoothing'] = np.average(np.concatenate([[data['globtemp'].shift(i) for i in range(5)]]), axis=0, weights=[stats.norm(0, 1).pdf(i) for i in range(5)])
plot = sns.lineplot(x='date', y='value', hue='variable', data=data.melt(id_vars='date'))
```

The k = 5 moving average smooths the series very well. The n = 5 kernel smoothing slightly smooths the sereis.
