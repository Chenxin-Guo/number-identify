### 3.1

In Example 3.5, $\rho(1) = \frac{\theta}{1 + \theta^2}$.

Because for any $\theta$, $(\theta + 1)^2$ must be non-negative, that is $(\theta - 1)^2 = \theta^2 - 2\theta + 1 \geq 0$. Therefore $\theta^2 + 1 \geq 2\theta$ and $\rho(1) = \frac{\theta}{1 + \theta^2} \geq \frac12$. Similarily, from the fact that $(\theta + 1)^2 \geq 0$ we can infer $\rho(1) \leq \frac12$.

When $\theta = 1$, $\rho(1)$ reaches its maximum value of $\frac12$, and when $\theta = -1$, $\rho(1)$ reaches its minimum value of $-\frac12$.

### 3.2

#### (a)

By induction, when $t=0$, $x_0 = \phi^0w_0$. 

Assume when $t=k$, $x_k = \Sigma_j \phi^jw_{k-j}$, then $x_{k+1} = \phi x_k + w_{k+1} = \phi \Sigma_j \phi^jw_{k-j} + w_{k+1} = \Sigma_j \phi^jw_{k+1-j}$. Therefore for any positive integer $t$, $x_t = \Sigma_j \phi^jw_{t-j}$.

#### (b)

$\mathbb{E}(x_t) = \Sigma_j \phi^j\mathbb{E}(w_{t-j}) = \Sigma_j \phi^j \cdot 0 = 0$

#### (c)

$var(x_t) = \mathbb{E}[(\Sigma_j^t \phi^jw_{t-j})(\Sigma_k^t \phi^kw_{t-k})]\\
=\Sigma_j^t \mathbb{E}(\phi^{2j}w^2_{t-j})\\
=\sigma_w^2 \Sigma_{j=0}^t(\phi^{2j}) = \frac{\sigma_w^2}{1 - \phi^2}(1 - \phi^{2(t+1)})$

#### (d)

$cov(x_{t+h}, x_t) = \mathbb{E}[(\Sigma_j^{t+h} \phi^jw_{t+h-j})(\Sigma_k^t \phi^kw_{t-k})]\\
=\mathbb{E}[(\Sigma_{k=0}^{t} \phi^{k+h}w_{t-k})(\Sigma_k^t \phi^kw_{t-k})]\\
=\phi^h\Sigma_k^t \mathbb{E}(\phi^{2k}w^2_{t-k})\\
=\phi^hvar(x_t)$

#### (e)

No, because $\gamma_x(s, t) = \phi^h var(x_t) = \frac{\phi^h\sigma_w^2}{1 - \phi^2}(1 - \phi^{2(t+1)})$, which depends on $t$.

#### (f)

When $t$ approach infinity, $\phi^{2(t+1)}→0$, so $\gamma_x(s, t) = \frac{\phi^h\sigma_w^2}{1 - \phi^2}(1 - \phi^{2(t+1)}) → \frac{\phi^h\sigma_w^2}{1 - \phi^2}$, which depends on $s, t$ only through $h = |s-t|$.

#### (g)

By multipling $\sigma_w^2$ to samples from $N(0, 1)$ we can get samples from $N(0, \sigma_w^2)$, which can be $w_t$, then we use $x_t = x_{t-1} + w_t$ to generate observations.

#### (h)

$x_t = \Sigma_{j=0}^{t-1} \phi^jw_{t-j} + \frac{\phi^tw_0}{\sqrt{1-\phi^2}}$, $\mathbb{E}(x_t) = 0$.

$var(x_t) = \mathbb{E}[(\Sigma_{j=0}^{t-1} \phi^jw_{t-j} + \frac{\phi^tw_0}{\sqrt{1-\phi^2}})(\Sigma_{j=0}^{t-1} \phi^jw_{t-j} + \frac{\phi^tw_0}{\sqrt{1-\phi^2}})]\\
=\Sigma_{j=0}^{t-1} \mathbb{E}(\phi^{2j}w^2_{t-j}) + \frac{\sigma_w^2\phi^{2t}}{{1-\phi^2}}\\
=\sigma_w^2 \Sigma_{j=0}^{t-1}(\phi^{2j}) + \frac{\sigma_w^2\phi^{2t}}{{1-\phi^2}}\\
= \frac{\sigma_w^2}{1 - \phi^2}(1 - \phi^{2t}) + \frac{\sigma_w^2\phi^{2t}}{{1-\phi^2}} = \frac{\sigma_w^2}{1 - \phi^2}$

, which is constant.

According to part (d), $\gamma_x(t+h, t) = \phi^h var(x_t) = \frac{\phi^h\sigma_w^2}{1 - \phi^2}$, which only depends on $h$. Hence $x_t$ is stationary.

### 3.3

#### (a)

Using $x_t = -\Sigma_{j=1}^{\infty}\phi^{-j}w_{t+j}, \mathbb{E}(x_t) = -\Sigma_{j=1}^{\infty}\phi^{-j}\mathbb{E}(w_{t+j}) = 0$, since $\Sigma_{j=1}^{\infty}\phi^{-j}$ converge when $|\phi| > 1$.

$\gamma_x(h) = \mathbb{E}[(-\Sigma_{j=1}^{\infty}\phi^{-j}w_{t+j})(-\Sigma_{k=1}^{\infty}\phi^{-k}w_{t+h+k})]\\
=\mathbb{E}[(\Sigma_{j=1}^{\infty}\phi^{-j}w_{t+j})(\Sigma_{j=h+1}^{\infty}\phi^{-j+h}w_{t+j})]\\
=\mathbb{E}[(\Sigma_{j=h+1}^{\infty}\phi^{-2j+h}w_{t+j})]\\
=\sigma_w^2\phi^{-h-2}\frac{1 - \phi^{-\infty}}{1-\phi^{-2}}\\
=\frac{\sigma_w^2\phi^{-h-2}}{1-\phi^{-2}}$

#### (b)

Since $|\phi| > 1$, $|\frac{1}{\phi}| < 1$. $y_t = \Sigma_{j=0}^{\infty}\phi^{-j}v_{t-j}$.

$\Sigma_{j=0}^{\infty}\phi^{-j} = \frac{1 - \phi^{-\infty}}{1 - \phi^{-1}} = \frac{\phi}{\phi - 1} < \infty$. So $y_t$ is casual.

$\mathbb{E}(y_t) = \mathbb{E}(\Sigma_{j=0}^{\infty}\phi^{-j}v_{t-j}) = \frac{\phi}{\phi - 1} \cdot 0 = 0$

$\gamma_y(h) = \mathbb{E}[(\Sigma_{j=0}^{\infty}\phi^{-j}v_{t-j})(\Sigma_{k=0}^{\infty}\phi^{-k}v_{t+h-k})]\\
=\mathbb{E}[(\Sigma_{j=0}^{\infty}\phi^{-j}v_{t-j})(\Sigma_{j=-h}^{\infty}\phi^{-j-h}v_{t-j})]\\
=\mathbb{E}[(\Sigma_{j=0}^{\infty}\phi^{-2j-h}v_{t-j})]\\
=\sigma_w^2\phi^{-2}\phi^{-h}\frac{1 - \phi^{-\infty}}{1-\phi^{-2}}\\
=\frac{\sigma_w^2\phi^{-h-2}}{1-\phi^{-2}} = \gamma_x(h)$

### 3.4

#### (a)

In operator form, $(1 - 0.8B + 0.15B^2)x_t = (1 - 0.3B)w_t$. But they have a common factor $1 - 0.3B$. 

$(1 - 0.3B)(1 - 0.5B)x_t = (1 - 0.3B)w_t$. After cancellation, $x_t = 0.5x_{t-1} + w_t$, which is an ARMA(1, 0) model. Since $\phi_1 < 1$, the model is casual. It is also invertible because $\theta(z) = 1$

#### (b)

In operator form, $(1 - B + 0.5B^2)x_t = (1 - B)w_t$, which is an ARMA(2, 1) model. The model is not casual because $\phi(z) = 1 - z + 0.5z^2 = 0$, when $z = 1 + i or 1 - i$. And the model is not invertible because $\theta(z) = 1 - z = 0$, when $z = 1$.

$|\phi_2| = |-\frac{1}{z_1z_2}| = |\frac{1}{z_1z_2}| < 1$.

$\phi_1 + \phi_2 = \frac{z_1 + z_2 -1}{z_1z_2} < \frac11 = 1$.

$\phi_2 - \phi_1 = \frac{-1 - z_1 - z_2}{z_1z_2} < \frac11 = 1$.

### 3.6

$\phi(z) = (1 + 0.9z^2)$, the root is $\pm\sqrt{\frac{10}{9}}i$, a complex conjugate pair. So $\rho(h) = a|z_1|^{-h}cos(h\theta + b)$. $\theta = tan^{-1}(\frac{\frac{10}{9}}{0}) = \frac{\pi}{2}$.

Since $\rho(0) = 1 = \alpha cos(b)$ and $\rho(-1) = \frac{\phi_1}{1 - \phi_2} = 0 = \alpha \sqrt{\frac{10}{9}}cos(-\frac{\pi}{2} + b)$. $a = 1, b = 0$, $\rho(h) = (\frac{10}{9})^{-h}cos(\frac{\pi h}{2})$

```python
rho = [math.sqrt(10/9)**(-h)*math.cos(h*math.pi/2) for h in range(100)]
plot = sns.lineplot(x = range(100), y=rho)
```

### 3.8

We can write the ARMA(1, 1) process in Example 3.14 as $(1 - \phi B)x_t = (1 + \theta)w_t$.

$x_t = \frac{1+\theta B}{1-\phi B}w_t = (1 + (\phi + \theta)B + (\phi^2 + \theta\phi^2)B^2 + \cdots)w_t = w_t + \Sigma_{j=1}^{\infty}[(\phi^j + \phi^{j-1}\theta)B]w_t\\
= w_t + \Sigma_{j=1}^{\infty}(\phi^j + \phi^{j-1}\theta)w_{t-j}$

$\gamma(0) = \mathbb{E}(w_tw_t) + \mathbb{E}[\Sigma_{j=1}^{\infty}(\phi^j + \phi^{j-1}\theta)(\phi^j + \phi^{j-1}\theta)w_{t-j}]\\
=(1 + (\phi + \theta)^2\frac{1 - \phi^{\infty}}{1 - \phi^2})\sigma_w^2 = \sigma_w^2\frac{1 + 2\theta\phi + \theta^2}{1 - \phi^2}$

$\gamma(h) = \mathbb{E}(w_tw_{t+h}) = \mathbb{E}[(w_{t+h} + \Sigma_{j=1}^{\infty}(\phi^j + \phi^{j-1}\theta)w_{t+h-j})(w_{t} + \Sigma_{k=1}^{\infty}(\phi^k + \phi^{k-1}\theta)w_{t-k})]\\
=\mathbb{E}[\Sigma_{j=1}^{\infty}(\phi^j + \phi^{j-1}\theta)w_{t+h-j}\Sigma_{j=1+h}^{\infty}(\phi^{j-h} + \phi^{j-h-1}\theta)w_{t+h-j}] + (\phi^h + \phi^{h-1}\theta)\sigma_w^2\\
=\Sigma_{j=1+h}^{\infty}(\phi^j + \phi^{j-1}\theta)(\phi^{j-h} + \phi^{j-h-1}\theta)w_{t+h-j}+ (\phi^h + \phi^{h-1}\theta)\sigma_w^2\\
= \sigma_w^2(\frac{(\phi^{1+h} + \phi^h\theta)(\phi+\theta)}{1-\phi^2} + \phi^h + \phi^{h-1}\theta)\\
= \sigma_w^2\frac{(1+\theta\phi)(\phi+\theta)}{1 - \phi^2}\phi^{h-1}$.

Here we get the same result as Example 3.14.

```python
phi, theta = 0.6, 0.9
arma_1_1 = [1]+[((1+theta*phi)*(theta+phi)*(phi**(h-1)))/(1+2*theta*phi+theta**2) for h in range(1, 10)]
arma_1_0 = [phi**h for h in range(10)]
arma_0_1 = [1, theta / (1+theta**2)] + [0] * 8
data = pd.DataFrame({'h':list(range(10))*3, 'acf':arma_1_1+arma_1_0+arma_0_1, 'model':['arma(1, 1)']*10 +['arma(1, 0)']*10 +['arma(0, 1)']*10})
plot = sns.lineplot(data=data, x='h', y='acf', hue='model')
```

$\rho_{arma(1, 1)}(h) \geq \rho_{arma(1, 0)}(h) \geq \rho_{arma(0, 1)}(h)$

```python
sample, acf, pacf = [], [], []
sample.append(['arma_1_1', tsa.arima_process.arma_generate_sample(ar=[1, -0.6], ma=[1, 0.9], nsample=100)])
sample.append(['arma_1_0', tsa.arima_process.arma_generate_sample(ar=[1, -0.6], ma=[1], nsample=100)])
sample.append(['arma_0_1', tsa.arima_process.arma_generate_sample(ar=[1], ma=[1, 0.9], nsample=100)])
for model, samples in sample:
    acf += list(tsa.stattools.acf(samples, nlags=9))
    pacf += list(tsa.stattools.pacf(samples, nlags=9))
sample_acf = pd.DataFrame({'h':list(range(10))*3, 'acf':acf, 'model':['acf arma(1, 1)']*10 +['acf arma(1, 0)']*10 +['acf arma(0, 1)']*10})
plot = sns.lineplot(data=pd.concat([data, sample_acf]), x='h', y='acf', hue='model')
```

```python
sample_pacf = pd.DataFrame({'h':list(range(10))*3, 'acf':pacf, 'model':['pacf arma(1, 1)']*10 +['pacf arma(1, 0)']*10 +['pacf arma(0, 1)']*10})
plot = sns.lineplot(data=pd.concat([sample_pacf, sample_acf]), x='h', y='acf', hue='model')
```

The pacf for ARMA(1, 0) and acf for ARMA(0, 1) cuts off after lag 1 while other statistics has a tails off behavior, which prefectly matches table 3.1.
