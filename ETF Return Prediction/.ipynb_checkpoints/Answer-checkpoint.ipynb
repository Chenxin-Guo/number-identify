3. close_return_adjust is the return adjusted for both dividends and splits.

### Q. 12

I used python package **pytrends** to automatically download the data. I choose the timeframe to be 'today 5-y', which refers to 2014 to present, and I set the geo to be 'US'.

```python
# Prepare stock list.
stock = pd.read_csv('./data/ETF_stocks.list', header=None)
stock_list = stock[0].tolist()
assert(len(stock_list) == 5)

stock_trend_path = './data/ETF_trend.csv'
if not os.path.exists(stock_trend_path):
    pytrends = TrendReq(hl='en-US', tz=360)
    pytrends.build_payload(stock_list, cat=0, timeframe='today 5-y', geo='US', gprop='')
    etf_trend = pytrends.interest_over_time()
    etf_trend = etf_trend.drop('isPartial', axis=1)
    etf_trend.to_csv(stock_trend_path)
    etf_trend = pd.read_csv(stock_trend_path, index_col='date')
else:
    etf_trend = pd.read_csv(stock_trend_path, index_col='date')
    
etf_trend.head()
```

### Q. 13

By observing the data, I notice that the some keywords, such as 'ASML.AS' does not have enough data. My explanation is that most people use 'ASML' instead of the full ticker as their search key word. So one solution is to use the brand name as the keyword instead. However, in that case, some companies such as 'Total' will collide the search for other entity. To handle that, I restricted Google Trends to only return results in the 'Business & Industrial' category.

```python
# Prepare stock list.
stock = pd.read_csv('./data/ETF_disambiguated.txt', header=None)
stock_list = stock[0].tolist()
assert(len(stock_list) == 5)

stock_trend_path = './data/ETF_trend_disambiguated.csv'
if not os.path.exists(stock_trend_path):
    etf_trend_disambiguated = []
    pytrends = TrendReq(hl='en-US', tz=360)
    # cat = 12 represents we only fetch result belongs to 'Business & Industrial' category.
    # full detail about the cat coding schema can be found at https://github.com/pat310/google-trends-api/wiki/Google-Trends-Categories
    pytrends.build_payload(stock_list, cat=12, timeframe='today 5-y', geo='US', gprop='')
    etf_trend_disambiguated = pytrends.interest_over_time()
    etf_trend_disambiguated = etf_trend_disambiguated.drop('isPartial', axis=1)
    etf_trend_disambiguated.to_csv(stock_trend_path)
    etf_trend_disambiguated = pd.read_csv(stock_trend_path, index_col='date')
else:
    etf_trend_disambiguated = pd.read_csv(stock_trend_path, index_col='date')
    
etf_trend_disambiguated.head()
```

### Q. 15

Although the Google Trend api can only return 5 years of weekly time in a single request, we can break our request into two parts. The first request scraps data from 2009-2014, and the second request scraps data from 2014-2019. I horizontally concated the result and verified that I indeed downloaded weekly data of the disambiguated version from 2009-11-22 to 2019-11-22.

```python
# Prepare stock list.
stock = pd.read_csv('./data/ETF_disambiguated.txt', header=None)
stock_list = stock[0].tolist()
assert(len(stock_list) == 5)

stock_trend_path = './data/ETF_trend_final.csv'
if not os.path.exists(stock_trend_path):
    etf_trend_disambiguated = []
    pytrends = TrendReq(hl='en-US', tz=360)
    # cat = 12 represents we only fetch result belongs to 'Business & Industrial' category.
    # full detail about the cat coding schema can be found at https://github.com/pat310/google-trends-api/wiki/Google-Trends-Categories
    pytrends.build_payload(stock_list, cat=12, timeframe='2014-11-21 2019-11-22', geo='US', gprop='')
    second_half = pytrends.interest_over_time()
    # Scrap data between 2009 to 2014
    pytrends.build_payload(stock_list, cat=12, timeframe='2009-11-21 2014-11-22', geo='US', gprop='')
    first_half = pytrends.interest_over_time()
    etf_trend_disambiguated = pd.concat([first_half, second_half])
    etf_trend_disambiguated = etf_trend_disambiguated.drop('isPartial', axis=1)
    etf_trend_disambiguated.to_csv(stock_trend_path)
    etf_trend_disambiguated = pd.read_csv(stock_trend_path, index_col='date')
else:
    etf_trend_disambiguated = pd.read_csv(stock_trend_path, index_col='date')
    
etf_trend_disambiguated.head()
```

### Q. 17

#### (a)

Personally I am in favor of converting the returns into weekly frequency for the following reason:

1. Daily data is too sensitive to noise. Using weekly data is more stable.
2. We have 5 years of data, which spans over ~260 weeks. That is enough data point to train a predictive model with an acceptable p-value.

#### (b)

I used Pandas's weekday() function to calculate the next sunday for each trade day. And then I can group by week and calculate the mean return over the week. I used the date of every Sunday as the index to align with Google trends data.

```python
from datetime import timedelta
trend = pd.read_csv(stock_trend_path, index_col='date')
trend.index.name = None
trend.index = pd.to_datetime(trend.index, format='%Y-%m-%d')
ret = pd.read_csv('./data/return.csv')
ret['Date'] = pd.to_datetime(ret['Date'], format='%Y-%m-%d')
ret['week'] = ret['Date'].apply(lambda x: x + timedelta(days = 6 - x.weekday()))
ret = ret.groupby(['week'])['SAP.DE', 'FP.PA', 'MC.PA', 'ASML.AS', 'LIN.DE'].mean()
ret.index.name = None
ret.columns = trend.columns
ret.head()
```

```python
df = pd.merge(ret, trend, left_index=True, right_index=True,  how='inner', suffixes=('_return', '_trend'))
plot_df = df.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
plot_df = plot_df.melt()
plot_df.index = list(df.index) * df.shape[1]
plot_df['source'] = plot_df['variable'].str.split('_').apply(lambda x: x[1])
plot_df['stock'] = plot_df['variable'].str.split('_').apply(lambda x: x[0])
plot_df['date'] = plot_df.index

g = sns.FacetGrid(plot_df, row='stock', hue='source', height=2.0, aspect=5)
g.map(sns.lineplot, 'date', 'value')
g.add_legend();
```

```python
for lag in range(-4, 5):
    print('For lag = %d' % lag)
    for col in ret.columns:
        temp = pd.DataFrame({'return': df[col+'_return'], 'trend':df[col+'_trend'].shift(lag)})
        temp.dropna(inplace=True)
        print('\tCorrelation of %s between trend and return rate is %lf' % (col, np.corrcoef(temp['return'], temp['trend'])[0][1]))
```

It is interesting to see that the return of individual stock and trend in the same week (|lag| <= 4) does not have strong correlation, which means that the search interest might not be a good indicator of the return of a stock in the near future.

### Q. 18

```python
etf = pd.read_csv('./data/fez.csv')
etf['Date'] = pd.to_datetime(etf['Date'], format='%Y-%m-%d')
etf['week'] = etf['Date'].apply(lambda x: x + timedelta(days = 6 - x.weekday()))
etf['return'] = etf['Close'].pct_change()
etf = etf.groupby(['week'])['return'].mean()
trend['return'] = etf
trend.dropna(inplace=True)
trend.head()
```

```python
for lag in range(6):
    trend['etf_lag_' + str(lag)] = trend['return'].shift(-lag)
    trend.dropna(inplace=True)
    print('The R-Square of lag %d is %lf' % (lag, sm.OLS(trend['etf_lag_' + str(lag)], trend[trend.columns[:5]]).fit().rsquared))
```

The R-Square for lag 0 to 5 are all small. The maximum of R^2 is when the lag is 2. Which means the search interest of the top 5 holding can be used to interpret 1% of variation in the change of ETF return two weeks later.
