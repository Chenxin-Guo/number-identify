### 1.6

#### (a)

$\mu_x(t) = \mathbb{E}(\beta_1 + \beta_2t + w_t) = \beta_1 + \beta_2t$, which depends on t. Therefore $x_t$ is not stationary.

#### (b)

$\mu_y(t) = \mathbb{E}(x_t - x_{t-1}) = \mathbb{E}(\beta_2 + w_t - w_{t-1}) = \beta_2$, which does not depend on t.

$\begin{align}
    \gamma_y(s, t) & = \mathbb{E}((x_t - x_{t-1} - \mu_y(t))(x_s - x_{s-1} - \mu_y(s)))\\
    & = \mathbb{E}(\beta_2 + w_t - w_{t-1} - \beta_2)(\beta_2 + w_s - w_{s-1} - \beta_2)\\
    & = \mathbb{E}(w_t - w_{t-1})(w_s - w_{s-1})\\
    & = \begin{cases} 
      0 & |s - t| \geq 2\\
      -\sigma_w^2 & |s - t| = 1\\
      2\sigma_w^2 & |s - t| = 0 \\
   \end{cases}
\end{align}$

, which only depends on |s - t|. According to Definition 1.7, this is weakly stationary time series.

#### (c)

$\mu_v(t) = \frac{1}{2q+1}\Sigma_{j=-q}^{q}\mathbb{E}(x_{t-j})=\frac{1}{2q+1}\Sigma_{j=-q}^{q}\mathbb{E}(\beta_1 + \beta_2(t-j)) = \frac{1}{2q+1}\frac{(2q+1)(\beta_1 + \beta_2(t - q) + \beta_1 + \beta_2(t+q))}{2} = \beta_1 + \beta_2t$

$\begin{align}
\gamma_v(s, t) & = \mathbb{E}((v(t) - \mu_v(t))(v(s) - \mu_v(s))\\
& = \mathbb{E}((\frac{\Sigma_{j=-q}^{q}w_{t-j}}{2q+1})(\frac{\Sigma_{j=-q}^{q}w_{s-j}}{2q+1}))\\
& = \mathbb{E}\frac{\Sigma_{j=-q}^{q}w_{t-j}\Sigma_{j=-q}^{q}w_{s-j}}{(2q+1)^2}\\
& = \begin{cases} 
      0 & |s - t| \geq 2q+1\\
      \frac{2q + 1 - |s - t|}{(2q+1)^2} & |s - t| < 2q+1\\
   \end{cases}
\end{align}$

### 1.7

$\mu_x(t) = \mathbb{E}(w_{t-1} + 2w_t + w_{t+1}) = 0$

$\gamma(s, t) = \mathbb{E}((w_{t-1} + 2w_t + w_{t+1})(w_{s-1} + 2w_s + w_{s+1}))$

If $|h| = |s - t| = 2$, $\gamma(s, t) = \mathbb{E}(w_iw_i) = 1$. If $|h| = |s - t| = 1$, $\gamma(s, t) = \mathbb{E}(2w_{i-2}w_{i-1} + 2w_{i}w_{i-1}) = 4$. If $s = t$, $\gamma(s, t) = \mathbb{E}(w_{i-1}w_{i-1} + 4w_{i}w_{i} + w_{i+1}w_{i+2}) = 6$.

$\gamma(s, t) = \begin{cases} 
      0 & |s - t| \geq 3\\ 
      1 & |s - t| = 2\\ 
      4 & |s - t| = 1\\ 
      6 & |s - t| = 0\\
   \end{cases}$
   
$\rho(s, t) = \frac{\gamma(s, t)}{\sqrt{\gamma(s, s)\gamma(t, t)}} = \begin{cases} 
      0 & |s - t| \geq 3\\ 
      \frac16 & |s - t| = 2\\ 
      \frac23 & |s - t| = 1\\ 
      1 & |s - t| = 0\\
   \end{cases}$

### 1.8

#### (a)

When $t=1$, $x_1 = \delta + w_1$. Assume when $t=n$, we have $x_n = \delta n + \Sigma_{k=1}^n w_k$, then $x_{n+1} = \delta + x_n + w_{n+1} = \delta (n+1) + \Sigma_{k=1}^{n+1} w_k$. 

By mathematical induction, for any positive integer t, $x_t = \delta t + \Sigma_{k=1}^t w_k$.

#### (b)

$\mu_x(t) = \mathbb{E}(\delta t + \Sigma_{k=1}^t w_k) = \delta t$

$\gamma_x(s, t) = \mathbb{E}((\Sigma_{k=1}^s w_k)(\Sigma_{k=1}^t w_k)) = min(s, t)\sigma_w^2$

#### (c)

$x_t$ is not stationary because $\mu_x(t)$ depends on $t$.

#### (d)

$\rho_x(t-1, t) = \frac{(t - 1)\sigma_w^2}{\sqrt{(t\sigma_w^2)((t-1)\sigma_w^2)}} = \sqrt{\frac{t-1}{t}}$.

As $t$ increase to infinity, $\rho_x(t-1, t)\rightarrow 1$, which means that the autocorrelation between two consecutive $t$ of a random walk approachs 1 when $t$ becomes larger.

#### (e)

Let $y_t = x_{t+1} - x_t - \delta = w_{t+1} - w_t$.

$\mu_y(t) = 0$, $\gamma_y(s, t) = \begin{cases} 
      0 & |s - t| \geq 2\\ 
      -1 & |s - t| = 1\\ 
      2 & |s - t| = 0\\
   \end{cases}$, which is stationary.

### 1.9

$\mu_x(t) = sin(2\pi w_0 t)\mathbb{E}(U_1) + cos(2\pi w_0 t)\mathbb{E}(U_2) = 0$.

$\gamma_x(s, t) = \mathbb{E}((sin(2\pi w_0 t)U_1 + cos(2\pi w_0 t)U_2)(sin(2\pi w_0 s)U_1 + cos(2\pi w_0 s)U_2))$.

Since $U_1, U_2$ are independent and $\mathbb{E}(U_1U_2) = 0$.

$\begin{align}
\gamma_x(s, t) &  = \mathbb{E}((sin(2\pi w_0 t)sin(2\pi w_0 s)U_1^2 + (cos(2\pi w_0 t)cos(2\pi w_0 s)U_2^2)\\
& = (sin(2\pi w_0 t)sin(2\pi w_0 s) + cos(2\pi w_0 t)cos(2\pi w_0 s))\sigma^2\\
& = \sigma^2cos(2\pi w_0 h)
\end{align}$.

By definition, $x_t$ is weakly stationary.

Let $\frac{dMSE(A)}{dA} = 0$, $A = \frac{\mathbb{E}(x_{t+l}x_t)}{\mathbb{E}(x_t^2)} = \frac{\gamma(t+l - t)}{\gamma{0}} = \rho(l)$.

#### (b)

=\gamma(0)\rho(l)^2 - 2\gamma(0)\rho(l)^2 + \gamma(0)\\
=\gamma(0)(1 - \rho(l)^2)$

#### (c)

### 1.13

   \end{cases}$

      0 & h \geq 2
   \end{cases}$
   
#### (b)

$\gamma_{xy}(h) = \mathbb{E}[(w_{t+h} + \theta w_{t+h-1} + u_{t+h})(w_t)] = \begin{cases}

   \end{cases}$
